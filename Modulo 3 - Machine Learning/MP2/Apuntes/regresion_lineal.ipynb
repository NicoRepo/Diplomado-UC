### Regresión Lineal Básica
Obtentremos el parametro *theta* con la siguente formula utilizando las:
$$ \theta = {(X^TX)^{-1} X^Ty} $$
*Nota: Se puede obtener usando el metodo del Mínimo error cuadrático medio*

```python
import numpy as np, matplotlib.pyplot as plt
np.random.seed(3)
#NOTE: Creamos un arreglo de datos al azar para variable dependiente
x = 3*np.random.rand(100)

#x
#NOTE: Creamos un arreglo de datos al azar para variable independientes
y = 5+2*x + np.random.randn(100, 1)
x_ = np.c_[np.ones((100,1)), x]
theta_opt = np.linalg.inv(x_.T.dot(x_)).dot(x_.T).dot(y)
theta_opt
```

### Grafico del modelo creado

```python
plt.plot(x, y, 'b.')
plt.ylabel('y')
plt.xlabel('X')
plt.axis([0,3,0,15])
plt.legend(["Regresión Lineal"])
plt.show()
```

### Realizar una predicción

```python
x_nuevo = np.array([[0], [3]])
_X = np.c_[np.ones((2,1)), x_nuevo]
y_pred = _X.dot(theta_opt)
y_pred
```

### Graficar la predicción

```python
plt.plot(x_nuevo, y_pred, 'r--')
plt.plot(x, y, "b.")
plt.ylabel('y')
plt.xlabel('X')
plt.axis([0,3,0,15])
plt.legend(["Regresión Lineal"])
plt.show()
```
